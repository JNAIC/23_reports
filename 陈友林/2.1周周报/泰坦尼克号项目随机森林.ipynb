```python
import numpy as np                         # 导入包
import pandas as pd
titanic = pd.read_csv('train.csv')         # 读取文件
test = pd.read_csv('test.csv')
# titanic.describe
```

```python
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())    # 缺失值填充
titanic['Embarked'] = titanic['Embarked'].fillna('S') 

test['Age'] = test['Age'].fillna(test['Age'].median())    # 缺失值填充
test['Embarked'] = test['Embarked'].fillna('S')
titanic.describe()
test.describe()
```

```python
titanic['Sex'].replace(to_replace = {'male':0, 'female':1},inplace = True)
titanic['Embarked'].replace(to_replace = {'S':0, 'C':1, 'Q':2},inplace = True)

test['Sex'].replace(to_replace = {'male':0, 'female':1},inplace = True)
test['Embarked'].replace(to_replace = {'S':0, 'C':1, 'Q':2},inplace = True)
# 热编码
x = titanic.drop(['Name','Ticket','Cabin','PassengerId'], axis = 1)       # 删除无关数据
to_test = test.drop(['Name','Ticket','Cabin','PassengerId'], axis = 1)
to_test
```

```python
# 准备训练集和测试集
from sklearn.model_selection import train_test_split
X = x.drop('Survived',axis=1)
y = titanic['Survived']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
# 数据标准化
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(X_train)
x_test = scaler.fit_transform(X_test)
al_test = scaler.fit_transform(to_test)
```

```python
# 训练随机森林模型
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 100, random_state = 42)
model.fit(x_train, y_train)
```

```python
# 模型评估
from sklearn.metrics import accuracy_score, classification_report
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"精确度:{accuracy:.2f}")
print(report)
```

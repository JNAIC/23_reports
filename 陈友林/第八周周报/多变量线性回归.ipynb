```python
# 导入函数
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv("D:\AI俱乐部\ex1data2.txt",sep=",",header = None,names = ['x1','x2','y'])
x1 = df['x1'] * 0.001
x2 = df['x2']
y = df['y'] * 0.00001
```

```python
# 个数获取及学习率、训练次数、变量初始值设置
m = len(x1)
learning_rate = 0.01
n = 3000
a = 1
b = 1
c = 1
```

```python
# 梯度下降找到损失函数最小值
for i in range(n):
    Y = a * x1 + b * x2 + c
    errors = Y - y
    a1 = (1 / m) * np.sum(errors * x1)
    b1 = (1 / m) * np.sum(errors * x2)
    c1 = (1 / m) * np.sum(errors)
    a -= (learning_rate / m) * np.sum(errors * x1)
    b -= (learning_rate / m) * np.sum(errors * x2) 
    c -= (learning_rate / m) * np.sum(errors)
```

```python
# 图像绘制
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
ax.scatter3D(x1,x2,y)
plt.show
x = np.linspace(min(x1),max(x1), 10)
y = np.linspace(min(x2),max(x2), 10)
x, y = np.meshgrid(x, y)
z = a * x + b * y + c
ax.plot_surface(x, y, z,alpha = 0.5)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
columns=['Population','profit']
city_data=pd.read_csv("D:/JNAIC/test_task/ex1data1.txt",names=columns)
city_data.head()
```

```python
x = city_data.Population
y = city_data.profit
plt.scatter(x,y)
plt.xlabel("population of city in 10.000s")
plt.ylabel("profit of city in 10.000s")
```

```python
def J(X,Y,alpha,theta1,theta2):
# def J(X,Y,alpha,theta1,theta2,iterations):
    J_history=[]
    i=0
    m=len(X)
    h=0
    lost=(1/m)*(Y[i]-h)*alpha
#     for i in range(iterations):
    while True:
        for i in range(m):
            h+=theta1+theta2*X[i]
            J_history.append(lost)
        theta1+=lost*1
        theta2+=lost*X[i]

        if -0.01<J_history[-1] < 0.01:
            break
    print(f"the finalily theta1 is{theta1},theta2 is {theta2}")
    return theta1,theta2
```

```python
# X=x
# Y=y
# theta1=1
# theta2=1
# alpha=0.01
# iterations=400
theta1,theta2=J(x,y,0.01,1,1)
```

```python
import pandas as pd
path="D:/vscode/vscode_python/pandas/winemag-data-130k-v2.csv"
reviews=pd.read_csv(path,index_col=0)
pd.set_option("display.max_rows", 5)
```

```python
"""使用rename()的方法去重新命名索引名称或者列的名称"""#使用python字典的形式
reviews.rename(columns={"points":"score"})
```

```python
"""实例2 ：将行的索引更换"""
reviews.rename(index={0:"firstEntry",1:"secondEntry"})
```

```python
"""使用rename_axis()方法可以在行索引与列索引之前添加索引"""
reviews.rename_axis("wines",axis="rows").rename_axis("fields",axis="columns")
```

```python
"""使用concat()方法将不同的DataFrame或Series结合起来"""
canadian_youtube=pd.read_csv("D:/vscode/vscode_python/pandas/archive/CAvideos.csv")
british_youtube=pd.read_csv("D:/vscode/vscode_python/pandas/archive/GBvideos.csv")
pd.concat([canadian_youtube,british_youtube])
```

```python
"""使用join()方法去结合"""
left=canadian_youtube.set_index(["title","trending_date"])
right=british_youtube.set_index(["title","trending_date"])
left.join(right,lsuffix="_CAN",rsuffix="_UK")
```

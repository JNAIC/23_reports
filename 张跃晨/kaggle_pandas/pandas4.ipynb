```python
import pandas as pd
path="D:/vscode/vscode_python/winemag-data-130k-v2.csv"
reviews=pd.read_csv(path,index_col=0)
pd.set_option("display.max_rows", 5)
```

```python
"""可以替代方法 value_counts()的操作"""
"""使用groupby()方法"""
reviews.groupby("points").points.count()
```

```python

"""利用groupby()去求该组中的最小值"""
reviews.groupby("points").price.min()
```

```python
"""groupby()与apply()的结合使用"""#不太懂
reviews.groupby("winery").apply(lambda df:df.title.iloc[0])
```

```python
"""按照多个列进行分组"""
reviews.groupby(["country","province"]).apply(lambda df:df.loc[df.points.idxmax()])
```

```python
"""利用agg()方法添加列的索引"""
reviews.groupby("country").price.agg([len,min,max])
```

```python
"""利用groupby实现多索引"""
reviews.groupby(["country","province"]).description.agg([len])
```

```python
"""查看多索引的类型"""
mi=reviews.groupby(["country","province"]).description.agg([len]).index
type(mi)
```

```python
"""将多索引转换为普通索引的方法可用 reset_index()"""
country_reviews=reviews.groupby(["country","province"]).description.agg([len])
country_reviews.reset_index()
```

```python
"""使用sort_values()对列的元素进行排列"""
country_reviews=country_reviews.reset_index()
country_reviews.sort_values(by="len")
```

```python
"""同样是使用sort_values()进行反向排列"""
country_reviews.sort_values(by="len",ascending=False)
```

```python
"""按索引值排列，使用sort_index()方法排列"""
country_reviews.sort_index()
```

```python
"""一次按多个列排序"""
country_reviews.sort_values(by=["country","len"])
```

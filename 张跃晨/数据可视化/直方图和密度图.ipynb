```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
iris_filepath="D:/vscode/vscode_python/data_visualization/data/iris.csv"
iris_data=pd.read_csv(iris_filepath,index_col="Id")
iris_data.head()
```

# 直方图

```python
#histplot函数可以创建直方图
```

```python
sns.histplot(iris_data["Petal Length (cm)"])
"""选择我们要绘制的列：Petal Length (cm)"""
```

# 密度图

```python
#使用kdeplot函数制作密度估计图（KDE）
```

```python
sns.kdeplot(iris_data["Petal Length (cm)"],fill=True)
"""fill=True用于着色"""
```

```python
#可以使用jointplot函数绘制二维(2D)KDE图
```

```python
sns.jointplot(x=iris_data["Petal Length (cm)"],y=iris_data["Sepal Width (cm)"],kind="kde")
"""除了中间的2Dkde之外，顶部还有x轴的kde图，右边还有y轴的kde图"""
```

# 颜色编码的绘图

```python
"""同样使用histplot函数来绘制，data用以获取数据，x设置需要包含的列的名称，hue用于将数据拆分成不同直方图的列"""
sns.histplot(data=iris_data,x="Petal Length (cm)",hue="Species")
plt.title("Histogram of Petal Lengths, by Species")
```

```python
"""可以使用kdeplot函数创建颜色编码的kde图，用法与上述一样"""
sns.kdeplot(data=iris_data,x="Petal Length (cm)",hue="Species",fill=True)
plt.title("Distribution of Petal Lengths, by Species")
```

图表的三个类型
1.Trend 趋势
  sns.line----折线图可以显示一段时间的趋势
  
2.Relatinship 关系
  sns.barplot----条形图可以比较不同组的数量关系
  sns.heatmap----热图可以在数字表中查找颜色编码关系
  sns.scatterplot----散点图可以显示两个连续变量的关系；使用颜色编码时，还可以显示第三个变量的关系
  sns.regplot----散点图中绘制回归线以显示两个变量的线性回归关系
  sns.lmplot----散点图中含有多个颜色编码的变量，可以显示多条线性回归线以查看线性关系
  sns.swarmplot----分类散点图显示连续变量和分类变量的关系

3.Distribute 分布
  sns.histplot----显示单个数值变量的分布
  sns.kdeplot----（二维）kde图显示单个（两个）数值变量的估计平滑分布
  sns.jointplot----显示2D kde图和每个单独变量的kde图

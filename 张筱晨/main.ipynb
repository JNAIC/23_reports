```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r'E:\智能系统俱乐部\23_sharing\week5\ex1data1.txt',names=['population','profit'])
```

```python
data.plot.scatter('population','profit',label='population')
plt.show()
```

```python
data.insert(0,'oness',1)
```

```python
X = data.iloc[:,0:-1]
y = data.iloc[:,-1]


X = X.values
y = y.values
y = y.reshape(97,1)
```

```python

#定义损失函数
def costF(X,y,theta):
    inner = np.power(X @ theta - y,2)
    return np.sum(inner)/(2*len(X))


```

```python
#theta的初始值
theta = np.zeros((2,1))
theta.shape
```

```python
cost_init = costF(X,y,theta)
print(cost_init)
```

```python
#定义梯度下降函数
def gradient(X,y,theta,alpha,iters):
    costs = []
    
    for i in range(iters):
        theta = theta - (X.T @ (X@theta - y)) * alpha / len(X)
        cost = costF(X,y,theta)
        costs.append(cost)
      
        
        if i % 10 == 0:
            print(cost)
            
            
    return theta,costs
```

```python
alpha = 0.02
iters = 1000

theta,costs = gradient(X,y,theta,alpha,iters)
```

```python
x = np.linspace(y.min(),y.max(),100)
y_= theta[0,0] + theta[1,0] * x


fig,ax = plt.subplots()
ax.scatter(X[:,1],y,label='training data')
ax.plot(x,y_,'r',label='predict')
ax.legend()

plt.show()
```

```python
from matplotlib import pyplot as plt
```

```python
x=range(2,26,2)
y=[15,13,14.5,17,20,25,26,26,24,22,18,15]
```

```python
fig=plt.figure(figsize=(20,8),dpi=80)#调整图片大小
plt.plot(x,y)
#绘制x轴
plt.xticks(x)
#plt.savefig()保存
plt.show()
```

```python
#下面是一个绘制气温的例子
import random
x=range(0,120)
y=[random.randint(20,35) for i in range(120)]


plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y)
#调整x轴的刻度
_xtick_labels=["10:{}".format(i) for i in range(60)]
_xtick_labels+=["11:{}".format(i) for i in range(60)]
plt.xticks(list(x)[::3],_xtick_labels[::3],rotation=45)#rotation旋转的度数
plt.show()
```

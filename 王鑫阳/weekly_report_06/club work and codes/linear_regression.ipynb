```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

data = pd.read_csv(r'C:\Users\NENOOD\Desktop\test_task\test_task\ex1data1.txt', header=None)
```

```python
X = data.iloc[:,0]
y = data.iloc[:,1]

sns.regplot(x=X,y=y)
```

```python
X.shape
```

```python
def line_reg(w, X, b):
    return np.dot(X, w)+b

def square_loss(y_hat, y):
    return ((y_hat - y)**2)/2

def update(epochs, lr, w, b):
    for epoch in range(epochs):
        y_hat = line_reg(w, X, b)
        w -= lr * np.sum((y_hat - y) * X) / 97
        b -= lr * np.sum(y_hat - y) / 97
        print(f"epoch:{epoch+1}  loss:{np.sum(square_loss(y_hat, y)/97)}")
        print(f"w':{lr * np.dot((y_hat - y).T, X)/97}  b':{lr * np.sum(y_hat - y)/97}\n")
    return w,b
```

```python
w = 1
b = 0.1
lr = 0.024
epochs = 500
w,b = update(epochs, lr, w, b)
y_hat = line_reg(w, X, b)
```
